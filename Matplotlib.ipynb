```
import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([8.5,8.9,9.35,9.19])
xpoints=np.array([2018,2019,2020,2021])
plt.plot(xpoints,ypoints)
plt.title("Grade Of B.tech")
plt.ylabel("Pointer")
plt.xlabel("Year Of semister")
plt.grid(color='red',linestyle=':',linewidth=0.5)
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 8 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.4, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([8.5,8.9,9.35,9.19])
xpoints=np.array([2018,2019,2020,2021])
plt.bar(xpoints,ypoints)
plt.title("Grade Of B.tech")
plt.ylabel("Pointer")
plt.xlabel("Year Of semister")
plt.grid(color='red',linestyle=':',linewidth=0.5)
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

y = np.array([30, 23, 18, 15, 9, 5])
mylabels = ["Fruit", "Proteins", "Vegetable", "Dairy","Grain","Other"]
myexplode = [0.1, 0.2, 0.1, 0.1 ,0.1,0.1]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.legend(title = "Diet Routeen")
plt.show() 
```
